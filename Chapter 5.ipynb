Gradient Descent with Multiple Inputs

```python
def neural_network(input,weights):
    out = 0
    for i in range(len(input)):
        out += input[i] * weights[i]
    return out

def ele_mul(delta,input):
    out = [0,0,0]
    for i in range(len(out)):
        out[i] += delta * input[i]
    return out
```

```python
toes = [0.5,9.5,9.9,9.0]
wlrec = [0.65,0.8,0.8,0.9]
nfans = [1.2,1.3,0.5,1.0]
alpha = 0.01
weights = [0.01,0.2,-0.1]
input = [toes[0],wlrec[0],nfans[0]]
win_or_lose = [1,1,0,1]
true = win_or_lose[0]

for i in range(1900):
    pred = neural_network(input,weights)
    delta = pred - true
    error = delta ** 2
    weight_delta = ele_mul(delta,input)
    
    print("Iteration: "+str(i+1))
    print("Pred: "+str(pred))
    print("Error: "+str(error))
    print("Delta: "+str(delta))
    print("Weights: "+str(weights))
    print("Weight Deltas: "+str(weight_delta))
    print("-----------------------")
    
    for i in range(len(weights)):
        weights[i] -= alpha * weight_delta[i]
```

Freezing One Weight

```python
toes = [0.5,9.5,9.9,9.0]
wlrec = [0.65,0.8,0.8,0.9]
nfans = [1.2,1.3,0.5,1.0]
alpha = 0.01
weights = [0.01,0.2,-0.1]
input = [toes[0],wlrec[0],nfans[0]]
win_or_lose = [1,1,0,1]
true = win_or_lose[0]

for i in range(1900):
    pred = neural_network(input,weights)
    delta = pred - true
    error = delta ** 2
    weight_delta = ele_mul(delta,input)
    weight_delta[0] = 0 #Freezing Weight
    
    print("Iteration: "+str(i+1))
    print("Pred: "+str(pred))
    print("Error: "+str(error))
    print("Delta: "+str(delta))
    print("Weights: "+str(weights))
    print("Weight Deltas: "+str(weight_delta))
    print("-----------------------")
    
    for i in range(len(weights)):
        weights[i] -= alpha * weight_delta[i]
```
